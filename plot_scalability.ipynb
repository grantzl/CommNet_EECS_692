```python
import numpy as np
import matplotlib.pyplot as plt

# open the file to read it
f = open("traffic_medium_agents_junc.txt", "r")

junctions = []
success = []
for line in f.readlines():
    num_junc, success_line = line.split(';')

    # Get the number of junctions
    num_junctions = int(num_junc)
    junctions.append(num_junctions)

    # Get the success
    success.append(float(success_line))

plt.plot(junctions, success)
plt.xlabel('Number of Junctions & Agents')
plt.ylabel('Model Success Rate')
plt.title('Medium Trained Model- Both Modified')


plt.savefig('Med_trained_success.png')

f.close()

```
